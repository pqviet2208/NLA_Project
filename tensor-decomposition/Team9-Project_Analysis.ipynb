```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_no_decompose = pd.read_csv("models/cifar10/vgg19/no_decompose/train_log.csv")
df_tucker_decompose = pd.read_csv("models/cifar10/vgg19/tucker_decompose/train_log.csv")
```

```python
df_no_decompose
```

```python
df_tucker_decompose
```

```python
df_tucker_decompose.columns
```

```python
plt.plot(df_no_decompose['cumulative_time'], df_no_decompose["test_top1_acc"], label="no_decompose")
plt.plot(df_tucker_decompose['cumulative_time'], df_tucker_decompose["test_top1_acc"], label="tucker_decompose")
plt.legend()
plt.xlabel(r"Train Time, c")
plt.title("Top 1 Accuracy")
plt.ylabel("Test Accuracy, %")
plt.show()
```

```python
plt.plot(df_no_decompose['cumulative_time'], df_no_decompose["train_top1_acc"], label="no_decompose")
plt.plot(df_tucker_decompose['cumulative_time'], df_tucker_decompose["train_top1_acc"], label="tucker_decompose")
plt.legend()
plt.xlabel(r"Train Time, c")
plt.title("Top 1 Accuracy")
plt.ylabel("Train Accuracy, %")
plt.show()
```
